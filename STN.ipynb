# 1. Get history of all scans from the API

```python
SITE_KEYS = ['added', 'domain', 'name']
SCAN_KEYS = ['grade', 'timestamp', 'live', 'valid_https', 
             'downgrades_https', 'defaults_to_https', 'hsts', 
             'hsts_max_age', 'hsts_entire_domain', 'hsts_preload_ready', 
             'hsts_preloaded', 'score']
```

```python
all_scans = []
```

```python
import time
```

```python
for index, row in main_results_df.iterrows():
    r = requests.get('{}?limit=10000'.format(row['all_scans']))
    time.sleep(2)  # Tis good to be gentle to our website
    df = pd.read_json(r.text)
    df = json_normalize(df['results'])
    for index, scan in df.iterrows():
        new_scan = {'added': row['added'],
             'domain': row['domain'],
             'name': row['name'],
             'grade': scan['results']['grade'],
             'timestamp': scan['results']['timestamp'],
             'live': scan['results']['live'],
             'valid_https': scan['results']['valid_https'],
             'downgrades_https': scan['results']['downgrades_https'],
             'defaults_to_https': scan['results']['defaults_to_https'],
             'hsts': scan['results']['hsts'],
             'hsts_max_age': scan['results']['hsts_max_age'],
             'hsts_entire_domain': scan['results']['hsts_entire_domain'],
             'hsts_preload_ready': scan['results']['hsts_preload_ready'],
             'hsts_preloaded': scan['results']['hsts_preloaded'],
             'score': scan['results']['score']
            }
        all_scans.append(new_scan)
```

# Slope Graph

A [slope graph](http://charliepark.org/slopegraphs/) is a perfect visualization for this kind of data: we want to show the evolution in score as a function of time for many news organizations, and we primarily care just about the start and stop values. 

First we need to take our all scan dataframe and pull out:

* Name of organization
* Initial score
* Final score

Probably the fastest way to do this is to sort all the data and then use `drop_duplicates` (sorting is fast, loops are slow):

```python
sorted_df = df_live_scans.sort_values(by='timestamp')

result_earliest_scan = sorted_df[['timestamp', 'name', 'score']].drop_duplicates('name', keep='first')
result_latest_scan = sorted_df[['timestamp', 'name', 'score']].drop_duplicates('name', keep='last')
```

Let's distill it down a bit:

```python
result_earliest_scan.columns = ['timestamp_first_scan', 'name', 'score_first_scan']
```

```python
result_latest_scan.columns = ['timestamp_latest_scan', 'name', 'score_latest_scan']
```

```python
result_earliest_scan.head()
```

```python
result_latest_scan.head()
```

```python
df_scan_evolution = result_earliest_scan.merge(result_latest_scan, on='name')
```

```python
news_orgs = df_scan_evolution['name'].values
cols = ['initial score', 'latest score']
df = pd.DataFrame(index=news_orgs, columns=cols)

df['initial score'] = df_scan_evolution['score_first_scan'].values
df['latest score'] = df_scan_evolution['score_latest_scan'].values
```

# Why are we using the max for this aggregation?

This is because we have some spurious scan results due to bugs like https://github.com/freedomofpress/securethenews/issues/151. These cause false low scores, so the mean is systematically biased low.  

```python
month_by_month_stats.transpose()
```

```python
month_by_month_stats.unstack()
```

```python
unstacked = month_by_month_stats.unstack()
```
